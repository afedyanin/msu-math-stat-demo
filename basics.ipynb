```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate Data
x = np.random.rand(50,)
y = x * 2 + np.random.normal(0, 0.3, 50)
z = np.random.rand(50,)

df = pd.DataFrame({
    "x": x,
    "y": y,
    "z": z
})
```

### Pearson's correlation
$r_{XY}=\frac{\sum^n_{i=1}{(X_i - \bar{X})(Y_i - \bar{Y})}}
{\sqrt{\sum^n_{i=1}{(X_i - \bar{X})^2}}\sqrt{\sum^n_{i=1}{(Y_i - \bar{Y})^2}}}$


```python
# Compute and show correlation matrix
corr_mat = df.corr()

plt.matshow(corr_mat)
plt.show()
```

```python
# Scatterplot
plt.scatter(df["x"], df["y"])
plt.show()
```
